**uACR(urine albumin-creatinine ratio)**

Aids in early detection of CKD as it shows whether you have albumin in your urine. Albumin is an essential protein that's normally found in the blood; it helps build muscle, repair tissue, and fight infection. But it should be in your blood, not your urine. When you have albumin (protein) in your urine, it is called albuminuria or proteinuria. One of the main jobs of your kidneys is to filter your blood; they retain important biomolecules such as protein, whilst excreting waste products like creatinine. If kidneys are healthy, they should let only very little protein go into your urine, or even none. But if your kidneys are damaged, protein can "leak" out of the kidneys into your urine. High albumin in urine is an early indicator of CKD, which can ultimately progress to kidney failure if left untreated. (link)

**Uric Acid**

Another biomarker that indicates kidney dysfunction. Uric acid promotes inflammation, oxidative stress and endothelial dysfunction which can lead to damage of renal vasculature and glomeruli, impairing filtration of blood. (link) It can also precede a decline in eGFR levels making it possible for even earlier detection of CKD onset. (link) Uric acid is produced by the breakdown of purines in the liver and is supposed to be filtered out of blood by kidneys. If kidney function declines, uric acid is not filtered effectively leading to build up in the blood (hyperuricemia). Hyperuricemia may also promote CKD progression. (link)

**BUN (Blood Urea Nitrogen)**

Measures the amount of nitrogen in the blood from urea, a waste product formed from protein metabolism. Protein consumed is broken down into amino acids which are then broken down in the liver for energy or storage, and removing nitrogen from the amino acids. Nitrogen is converted into ammonia (toxic) that is detoxified in the liver by being converted to urea (water soluble & safe to circulate in the blood). Urea then enters the blood and BUN measures the nitrogen portion of the urea present. Kidneys are supposed to filter urea out of the blood into urine. When kidney filtration function decreases, BUN levels increase due to urea accumulating in blood. Normal BUN levels are 7‚Äì20 mg/dL but can differ slightly. Higher levels = kidney dysfunction, dehydration, high protein intake. Lower levels = liver disease, malnutrition, overhydration. (link)

#**Merging Datasets (2007-2018)**

```python
# -----------------------------
# üì¶ Import Required
# -----------------------------
import pandas as pd
import numpy as np

# -----------------------------
# Define Years and Cycles (2007‚Äì2018 only)
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

# -----------------------------
# Unified Loader Function (Preserves SEQN, Skips Missing Columns)
# -----------------------------
def load_data(file_prefix, columns):
    df_list = []
    for year, cycle in zip(years, cycles):
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{file_prefix}_{cycle}.xpt"
        try:
            df = pd.read_sas(url, format='xport')
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df_list.append(df[["SEQN"] + existing_cols])
        except Exception as e:
            print(f"Error loading {url}: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        return pd.DataFrame(columns=["SEQN"] + columns)

# -----------------------------
# Load Datasets
# -----------------------------
albcr_df = load_data("ALB_CR", ["URXUMA", "URXUCR"])
biopro_df = load_data("BIOPRO", ["LBDSBUSI"])
uas_df = load_data("UAS", ["URXUAC"])
demo_basic = load_data("DEMO", ["RIDAGEYR", "RIAGENDR", "RIDRETH3"])
demo_income = load_data("DEMO", ["INDFMIN2"])
bmi_df = load_data("BMX", ["BMXBMI"])
paq_df = load_data("PAQ", ["PAQ650"])  # Optional
fs_df = load_data("FSQ", ["FSDHH"])

# -----------------------------
# Merge All on SEQN
# -----------------------------
merged_df = albcr_df \
    .merge(biopro_df, on="SEQN", how="outer") \
    .merge(uas_df, on="SEQN", how="outer") \
    .merge(demo_basic, on="SEQN", how="outer") \
    .merge(demo_income, on="SEQN", how="outer") \
    .merge(bmi_df, on="SEQN", how="outer") \
    .merge(paq_df, on="SEQN", how="outer") \
    .merge(fs_df, on="SEQN", how="outer")

# -----------------------------
# Filter: Participants Aged 20 and Above
# -----------------------------
age_columns = [col for col in merged_df.columns if "RIDAGEYR" in col]
merged_df["RIDAGEYR"] = merged_df[age_columns].bfill(axis=1).iloc[:, 0]
merged_df = merged_df[merged_df["RIDAGEYR"] >= 20]

# -----------------------------
# Compute uACR
# -----------------------------
merged_df["uACR"] = merged_df["URXUMA"] / merged_df["URXUCR"]

# -----------------------------
# Preview
# -----------------------------
print("‚úÖ Final merged dataset preview:")
print(merged_df.head())
print("\nMissing values summary:")
print(merged_df.isnull().sum())
```

**Data Cleaning**

To ensure reliable analysis, rows with missing values were removed for key variables and biomarkers.

```python
# -------------------------------
# üìã Variables used in your study
# -------------------------------
columns = [
    "SEQN",
    "URXUMA",     # Urinary albumin
    "URXUCR",     # Urinary creatinine
    "URXUAC",     # Uric acid
    "LBDSBUSI",   # Blood urea nitrogen
    "RIDAGEYR",   # Age
    "RIAGENDR",   # Gender
    "RIDRETH3",   # Race/Ethnicity
    "INDFMIN2",   # Family income-to-poverty ratio
    "BMXBMI",     # BMI
    "PAQ650",     # Physical activity
    "FSDHH"       # Food security
]

# -------------------------------
# üßº 1. Drop rows with any missing values in these columns
# -------------------------------
merged_df.dropna(subset=columns, inplace=True)

# -------------------------------
# üö´ 2. Remove invalid codes from categorical variables
# -------------------------------
invalid_codes = [7, 9, 77, 99, 100]

merged_df = merged_df[
    ~merged_df["PAQ650"].isin(invalid_codes) &
    ~merged_df["FSDHH"].isin(invalid_codes) &
    ~merged_df["INDFMIN2"].isin(invalid_codes)
]

# -------------------------------
# üßº 3. Filter biologically implausible or top-coded values
# -------------------------------

# Age: Keep only adults ‚â• 20 years old
merged_df = merged_df[merged_df["RIDAGEYR"] >= 20]

# INDFMIN2: Keep only plausible ratios (e.g., ‚â§ 5.0)
merged_df = merged_df[merged_df["INDFMIN2"] <= 5.0]

# BMI: Filter implausible values (<10 or >70)
merged_df = merged_df[(merged_df["BMXBMI"] >= 10) & (merged_df["BMXBMI"] <= 70)]

# UACR-related: Avoid division by zero for URXUCR later on
merged_df = merged_df[merged_df["URXUCR"] > 0]

# -------------------------------
# üîÑ Optional: Reset index
# -------------------------------
merged_df.reset_index(drop=True, inplace=True)

# -------------------------------
# ‚úÖ Preview cleaned dataset
# -------------------------------
print("‚úÖ Rows remaining after full cleaning:", merged_df.shape[0])
print(merged_df.head())

```

**Variable Mapping**

To improve interpretability, categorical codes for gender and ethnicity were converted into readable labels using mapping dictionaries. This step enhances clarity for analysis and visualization.

```python
# -----------------------------
# üßÆ Define Categorical and Numerical Columns
# -----------------------------
cat_cols = ["RIAGENDR", "RIDRETH3", "PAQ650", "FSDHH"]
num_cols = ["URXUMA", "URXUCR", "URXUAC", "LBDSBUSI", "RIDAGEYR", "BMXBMI", "INDFMIN2"]

# -----------------------------
# üó∫ Mapping Dictionaries
# -----------------------------
gender_map = {
    1: "Male",
    2: "Female"
}

ethnicity_map = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other Race - Including Multi-Racial"
}

# -----------------------------
# üîç Check Original Values Before Mapping
# -----------------------------
print("Original RIAGENDR values:", merged_df["RIAGENDR"].unique())
print("Original RIDRETH3 values:", merged_df["RIDRETH3"].unique())

# -----------------------------
# üîÅ Apply Mappings with Fallback
# -----------------------------
merged_df["RIAGENDR"] = merged_df["RIAGENDR"].map(gender_map).fillna("Unknown")
merged_df["RIDRETH3"] = merged_df["RIDRETH3"].map(ethnicity_map).fillna("Other / Unknown")

# -----------------------------
# ‚úÖ Confirm Mapping Worked
# -----------------------------
print("Mapped Gender categories:", merged_df["RIAGENDR"].unique())
print("Mapped Ethnicity categories:", merged_df["RIDRETH3"].unique())
```

```python
# -----------------------------
# üó∫ Additional Mapping Dictionaries
# -----------------------------
fsdhh_map = {
    1: "Full Food Security",
    2: "Marginal Food Security",
    3: "Low Food Security",
    4: "Very Low Food Security"
}

paq650_map = {
    1: "Yes",
    2: "No"
}

# -----------------------------
# üîÅ Apply Additional Mappings
# -----------------------------
merged_df["FSDHH"] = merged_df["FSDHH"].map(fsdhh_map).fillna("Unknown")
merged_df["PAQ650"] = merged_df["PAQ650"].map(paq650_map).fillna("Unknown")

# Create Income Group from INDFMIN2
def bin_income(pir):
    if pir < 2:
        return "Low Income"
    elif pir < 4:
        return "Moderate Income"
    else:
        return "High Income"

merged_df["IncomeGroup"] = merged_df["INDFMIN2"].apply(bin_income)

# Optional: Set ordered categories for IncomeGroup
merged_df["IncomeGroup"] = pd.Categorical(
    merged_df["IncomeGroup"],
    categories=["Low Income", "Moderate Income", "High Income"],
    ordered=True
)

# -----------------------------
# ‚úÖ Confirm Mappings Worked
# -----------------------------
print("Mapped Food Security categories:", merged_df["FSDHH"].unique())
print("Mapped Physical Activity categories:", merged_df["PAQ650"].unique())
print("Binned Income Groups:", merged_df["IncomeGroup"].unique())
```

- **Normality Test ‚Äì Shapiro-Wilk**

To test if variables were normally distributed, the Shapiro-Wilk test was applied. However, this test requires that:

- Sample size must be between **3 and 5000**
- Variables must contain sufficient non-missing values

As some variables had more than 5000 valid samples, we randomly sampled **5000 rows** for each testable variable. Variables with zero valid samples were skipped and manually corrected.

```python
from scipy.stats import shapiro

# List of numerical columns
num_cols = ["uACR", "URXUAC", "LBDSBUSI", "RIDAGEYR", "BMXBMI"]

# Perform Shapiro-Wilk test and print results
print("Shapiro-Wilk Normality Test Results:")
for col in num_cols:
    data = merged_df[col].dropna()
    if len(data) > 5000:
        data = data.sample(5000, random_state=1)  # Shapiro test limit
    stat, p = shapiro(data)
    print(f"{col}: W = {stat:.4f}, p = {p:.4e} -> {'Not Normal' if p < 0.05 else 'Normal'}")
```

**Interpretation:**  
All consolidated continuous variables showed p-values < 0.05 in the Shapiro-Wilk test, indicating significant deviation from normality, which is expected for biological data.
Thus, non-parametric statistical tests are more appropriate for subsequent analyses.

```python
# Mapping dictionaries
indfmin2_map = {
    1.0: "PIR < 1.0", 2.0: "PIR 1.0‚Äì1.99", 3.0: "PIR 2.0‚Äì2.99",
    4.0: "PIR 3.0‚Äì3.99", 5.0: "PIR 4.0‚Äì4.99", 6.0: "PIR ‚â• 5.0",
    7.0: "PIR Very High", 8.0: "PIR Very High", 9.0: "PIR Very High",
    10.0: "PIR Very High", 12.0: "PIR Very High", 13.0: "PIR Very High",
    14.0: "PIR Very High", 15.0: "PIR Very High"
}
fsdhh_map = {1.0: "Very hard", 2.0: "Somewhat hard", 3.0: "Not very hard", 4.0: "Not hard at all"}
paq650_map = {1.0: "Yes (Active)", 2.0: "No (Inactive)"}

# Label mapping function
def map_label(val, var_name):
    if var_name == "INDFMIN2": return indfmin2_map.get(val, f"Group {val}")
    if var_name == "FSDHH": return fsdhh_map.get(val, f"Group {val}")
    if var_name == "PAQ650": return paq650_map.get(val, f"Group {val}")
    return f"Group {val}"

# Kruskal + Dunn function
def run_labeled_kruskal_dunn(df, group_col, val_col='uACR', exclude_vals=[77, 99, 9]):
    df_clean = df[[group_col, val_col]].dropna()
    df_clean = df_clean[~df_clean[group_col].isin(exclude_vals)]

    print(f"\nüîç Kruskal‚ÄìWallis H Test for {group_col}")
    groups = [g[val_col].dropna() for _, g in df_clean.groupby(group_col)]
    h_stat, p_val = kruskal(*groups)
    print(f"‚Üí H = {h_stat:.2f}, p = {p_val:.4f}")

    if p_val < 0.05:
        print("‚úÖ Significant difference detected. Key pairwise differences:\n")
        dunn_result = sp.posthoc_dunn(df_clean, val_col=val_col, group_col=group_col, p_adjust='bonferroni')

        count = 0
        for i in range(len(dunn_result)):
            for j in range(i + 1, len(dunn_result)):
                p = dunn_result.iloc[i, j]
                if p < 0.05:
                    g1, g2 = dunn_result.index[i], dunn_result.columns[j]
                    print(f"‚Ä¢ {map_label(g1, group_col)} vs {map_label(g2, group_col)}: p = {p:.4f}")
                    count += 1
        if count == 0:
            print("No significant pairwise differences after Bonferroni correction.")
    else:
        print("‚ùå No significant differences between groups.")

# Run for variables
run_labeled_kruskal_dunn(clean_df, 'INDFMIN2', exclude_vals=[77, 99])
run_labeled_kruskal_dunn(clean_df, 'FSDHH', exclude_vals=[77, 99])
run_labeled_kruskal_dunn(clean_df, 'PAQ650', exclude_vals=[9])
```

```python
from scipy.stats import mannwhitneyu, kruskal
import scikit_posthocs as sp

# Debug: check available columns
print("Available columns:", clean_df.columns.tolist())

# 1Ô∏è‚É£ Create CKD levels if not present
if 'CKD_level' not in clean_df.columns:
    def classify_ckd(uacr):
        if uacr < 30:
            return "Normal"
        elif 30 <= uacr < 300:
            return "Microalbuminuria"
        else:
            return "Macroalbuminuria"

    clean_df['CKD_level'] = clean_df['uACR'].apply(classify_ckd)

print("\n‚úÖ CKD levels created. Distribution:")
print(clean_df['CKD_level'].value_counts())

# 2Ô∏è‚É£ Annual change in uACR using YEAR column
if 'YEAR' in clean_df.columns:
    print("\nüîç Kruskal‚ÄìWallis Test for Annual uACR Changes")
    year_groups = [g['uACR'].dropna() for _, g in clean_df.groupby('YEAR')]
    print("Year group sizes:", [len(g) for g in year_groups])

    if len(year_groups) > 1:
        h_stat, p_val = kruskal(*year_groups)
        print(f"H = {h_stat:.2f}, p = {p_val:.4f}")

        if p_val < 0.05:
            print("‚úÖ Significant difference detected. Running Dunn's post-hoc test...")
            dunn_result = sp.posthoc_dunn(clean_df, val_col='uACR', group_col='YEAR', p_adjust='bonferroni')
            print(dunn_result)
        else:
            print("‚ùå No significant annual differences in uACR.")
    else:
        print("‚ö†Ô∏è Only one year group found. Skipping test.")
else:
    print("\n‚ö†Ô∏è 'YEAR' column not found. Skipping annual test.")
# 3Ô∏è‚É£ CKD vs categorical variables
for var in ['INDFMIN2', 'FSDHH', 'PAQ650']:
    print(f"\nüîç CKD level comparison by {var}")

    # Drop invalid values
    df_var = clean_df[[var, 'CKD_level']].dropna()
    if df_var[var].nunique() > 2:  # Kruskal‚ÄìWallis
        groups = [g['CKD_level'].astype('category').cat.codes for _, g in df_var.groupby(var)]
        if len(groups) > 1:
            h_stat, p_val = kruskal(*groups)
            print(f"H = {h_stat:.2f}, p = {p_val:.4f}")
            if p_val < 0.05:
                print("‚úÖ Significant difference detected. Dunn's post-hoc:")
                dunn_result = sp.posthoc_dunn(df_var, val_col='CKD_level', group_col=var, p_adjust='bonferroni')
                print(dunn_result)
            else:
                print("‚ùå No significant difference.")
        else:
            print("‚ö†Ô∏è Not enough groups for Kruskal‚ÄìWallis.")
    else:  # Mann‚ÄìWhitney for binary vars
        unique_vals = df_var[var].unique()
        if len(unique_vals) == 2:
            g1 = df_var[df_var[var] == unique_vals[0]]['CKD_level'].astype('category').cat.codes
            g2 = df_var[df_var[var] == unique_vals[1]]['CKD_level'].astype('category').cat.codes
            stat, p_val = mannwhitneyu(g1, g2, alternative='two-sided')
            print(f"Mann‚ÄìWhitney U: U = {stat:.0f}, p = {p_val:.4f}")
        else:
            print("‚ö†Ô∏è Not enough groups for Mann‚ÄìWhitney.")
```
