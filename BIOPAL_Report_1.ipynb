**uACR(urine albumin-creatinine ratio)**

Aids in early detection of CKD as it shows whether you have albumin in your urine. Albumin is an essential protein that's normally found in the blood; it helps build muscle, repair tissue, and fight infection. But it should be in your blood, not your urine. When you have albumin (protein) in your urine, it is called albuminuria or proteinuria. One of the main jobs of your kidneys is to filter your blood; they retain important biomolecules such as protein, whilst excreting waste products like creatinine. If kidneys are healthy, they should let only very little protein go into your urine, or even none. But if your kidneys are damaged, protein can "leak" out of the kidneys into your urine. High albumin in urine is an early indicator of CKD, which can ultimately progress to kidney failure if left untreated. (link)

**Uric Acid**

Another biomarker that indicates kidney dysfunction. Uric acid promotes inflammation, oxidative stress and endothelial dysfunction which can lead to damage of renal vasculature and glomeruli, impairing filtration of blood. (link) It can also precede a decline in eGFR levels making it possible for even earlier detection of CKD onset. (link) Uric acid is produced by the breakdown of purines in the liver and is supposed to be filtered out of blood by kidneys. If kidney function declines, uric acid is not filtered effectively leading to build up in the blood (hyperuricemia). Hyperuricemia may also promote CKD progression. (link)

**BUN (Blood Urea Nitrogen)**

Measures the amount of nitrogen in the blood from urea, a waste product formed from protein metabolism. Protein consumed is broken down into amino acids which are then broken down in the liver for energy or storage, and removing nitrogen from the amino acids. Nitrogen is converted into ammonia (toxic) that is detoxified in the liver by being converted to urea (water soluble & safe to circulate in the blood). Urea then enters the blood and BUN measures the nitrogen portion of the urea present. Kidneys are supposed to filter urea out of the blood into urine. When kidney filtration function decreases, BUN levels increase due to urea accumulating in blood. Normal BUN levels are 7‚Äì20 mg/dL but can differ slightly. Higher levels = kidney dysfunction, dehydration, high protein intake. Lower levels = liver disease, malnutrition, overhydration. (link)

```python
# -----------------------------
# üì¶ Import Required
# -----------------------------
import pandas as pd
import numpy as np

# -----------------------------
# Define Years and Cycles (2007‚Äì2018 only)
# -----------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["E", "F", "G", "H", "I", "J"]

# -----------------------------
# Unified Loader Function (Preserves SEQN, Skips Missing Columns)
# -----------------------------
def load_data(file_prefix, columns):
    df_list = []
    for year, cycle in zip(years, cycles):
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{file_prefix}_{cycle}.xpt"
        try:
            df = pd.read_sas(url, format='xport')
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df_list.append(df[["SEQN"] + existing_cols])
        except Exception as e:
            print(f"Error loading {url}: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        return pd.DataFrame(columns=["SEQN"] + columns)

# -----------------------------
# Load Datasets
# -----------------------------
albcr_df = load_data("ALB_CR", ["URXUMA", "URXUCR"])
biopro_df = load_data("BIOPRO", ["LBDSBUSI"])
uas_df = load_data("UAS", ["URXUAC"])
demo_basic = load_data("DEMO", ["RIDAGEYR", "RIAGENDR", "RIDRETH3"])
demo_income = load_data("DEMO", ["INDFMIN2"])
bmi_df = load_data("BMX", ["BMXBMI"])
paq_df = load_data("PAQ", ["PAQ650"])  # Optional
fs_df = load_data("FSQ", ["FSDHH"])

# -----------------------------
# Merge All on SEQN
# -----------------------------
merged_df = albcr_df \
    .merge(biopro_df, on="SEQN", how="outer") \
    .merge(uas_df, on="SEQN", how="outer") \
    .merge(demo_basic, on="SEQN", how="outer") \
    .merge(demo_income, on="SEQN", how="outer") \
    .merge(bmi_df, on="SEQN", how="outer") \
    .merge(paq_df, on="SEQN", how="outer") \
    .merge(fs_df, on="SEQN", how="outer")

# -----------------------------
# Filter: Participants Aged 20 and Above
# -----------------------------
age_columns = [col for col in merged_df.columns if "RIDAGEYR" in col]
merged_df["RIDAGEYR"] = merged_df[age_columns].bfill(axis=1).iloc[:, 0]
merged_df = merged_df[merged_df["RIDAGEYR"] >= 20]

# -----------------------------
# Compute uACR
# -----------------------------
merged_df["uACR"] = merged_df["URXUMA"] / merged_df["URXUCR"]

# -----------------------------
# Preview
# -----------------------------
print("‚úÖ Final merged dataset preview:")
print(merged_df.head())
print("\nMissing values summary:")
print(merged_df.isnull().sum())
```

```python
# Categorize uACR (URXUMA/URXUCR)
def categorize_uacr(row):
    if row["URXUCR"] == 0 or pd.isna(row["URXUMA"]) or pd.isna(row["URXUCR"]):
        return np.nan
    acr = row["URXUMA"] / row["URXUCR"]
    if acr < 30:
        return "Normal"
    elif acr < 300:
        return "Moderately increased"
    else:
        return "Severely increased"

# Categorize Uric Acid (URXUAC)
def categorize_uric_acid(val):
    if pd.isna(val):
        return np.nan
    elif val < 3.5:
        return "Low"
    elif val <= 7.2:
        return "Normal"
    else:
        return "High"

# Categorize BUN (LBDSBUSI)
def categorize_bun(val):
    if pd.isna(val):
        return np.nan
    elif val < 7:
        return "Low"
    elif val <= 20:
        return "Normal"
    else:
        return "High"

# Apply categorization
merged_df["uACR_Category"] = merged_df.apply(categorize_uacr, axis=1)
merged_df["UricAcid_Category"] = merged_df["URXUAC"].apply(categorize_uric_acid)
merged_df["BUN_Category"] = merged_df["LBDSBUSI"].apply(categorize_bun)
```

```python
def load_data(file_prefix, columns):
    df_list = []
    for year, cycle in zip(years, cycles):
        url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{file_prefix}_{cycle}.xpt"
        try:
            df = pd.read_sas(url, format='xport')
            existing_cols = [col for col in columns if col in df.columns]
            if existing_cols:
                df = df[["SEQN"] + existing_cols].copy()
                df["YEAR"] = f"{year}-{str(int(year)+1)}"  # Add year range label like '2007‚Äì2008'
                df_list.append(df)
        except Exception as e:
            print(f"Error loading {url}: {e}")
    if df_list:
        return pd.concat(df_list, ignore_index=True)
    else:
        return pd.DataFrame(columns=["SEQN"] + columns + ["YEAR"])

# Re-load datasets using the updated load_data function
albcr_df = load_data("ALB_CR", ["URXUMA", "URXUCR"])
biopro_df = load_data("BIOPRO", ["LBDSBUSI"])
uas_df = load_data("UAS", ["URXUAC"])
demo_basic = load_data("DEMO", ["RIDAGEYR", "RIAGENDR", "RIDRETH3"]) # Load RIDRETH3 here
demo_income = load_data("DEMO", ["INDFMIN2"])
bmi_df = load_data("BMX", ["BMXBMI"])
paq_df = load_data("PAQ", ["PAQ650"])  # Optional
fs_df = load_data("FSQ", ["FSDHH"])

# Merge all on SEQN and YEAR
merged_df = albcr_df \
    .merge(biopro_df, on=["SEQN", "YEAR"], how="outer") \
    .merge(uas_df, on=["SEQN", "YEAR"], how="outer") \
    .merge(demo_basic, on=["SEQN", "YEAR"], how="outer") \
    .merge(demo_income, on=["SEQN", "YEAR"], how="outer") \
    .merge(bmi_df, on=["SEQN", "YEAR"], how="outer") \
    .merge(paq_df, on=["SEQN", "YEAR"], how="outer") \
    .merge(fs_df, on=["SEQN", "YEAR"], how="outer")

# Apply mappings for categorical variables *after* merging
gender_map = {
    1: "Male",
    2: "Female"
}

ethnicity_map = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other Race - Including Multi-Racial"
}

merged_df["RIAGENDR"] = merged_df["RIAGENDR"].map(gender_map)
merged_df["RIDRETH3"] = merged_df["RIDRETH3"].map(ethnicity_map)

# Drop rows with NaN in RIAGENDR after mapping
merged_df.dropna(subset=["RIAGENDR"], inplace=True)


# Filter: Participants Aged 20 and Above (and below 65 as per step 6)
merged_df = merged_df[(merged_df["RIDAGEYR"] >= 20) & (merged_df["RIDAGEYR"] <= 64)].copy()

# Compute uACR
merged_df["uACR"] = merged_df["URXUMA"] / merged_df["URXUCR"]


# Check how many participants per survey cycle
print("‚úÖ Participant counts by NHANES cycle year:")
print(merged_df["YEAR"].value_counts().sort_index())

import matplotlib.pyplot as plt

# Count participants per year
year_counts = merged_df["YEAR"].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
year_counts.plot(kind="bar", edgecolor='black')
plt.title("Participant Count by NHANES Cycle Year")
plt.xlabel("Survey Cycle (YEAR)")
plt.ylabel("Number of Participants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# Step 7: Socioeconomic Categorization
# -----------------------------
def categorize_income(pir):
    if pd.isnull(pir): return np.nan
    elif pir <= 1.0:
        return "At or below poverty"
    elif pir <= 2.0:
        return "Low income"
    elif pir <= 3.5:
        return "Middle income"
    elif pir < 5.0:
        return "Upper middle income"
    else:
        return "High income (capped)"

merged_df['SES_Category'] = merged_df['INDFMIN2'].apply(categorize_income)

```

```python
# Categorize uACR (URXUMA/URXUCR)
def categorize_uacr(row):
    if row["URXUCR"] == 0 or pd.isna(row["URXUMA"]) or pd.isna(row["URXUCR"]):
        return np.nan
    acr = row["URXUMA"] / row["URXUCR"]
    if acr < 30:
        return "Normal"
    elif acr < 300:
        return "Moderately increased"
    else:
        return "Severely increased"

# Apply categorization
merged_df["uACR_Category"] = merged_df.apply(categorize_uacr, axis=1)
```

```python
# Categorize Uric Acid (URXUAC)
def categorize_uric_acid(val):
    if pd.isna(val):
        return np.nan
    elif val < 3.5:
        return "Low"
    elif val <= 7.2:
        return "Normal"
    else:
        return "High"

# Apply categorization
merged_df["UricAcid_Category"] = merged_df["URXUAC"].apply(categorize_uric_acid)
```

```python
# Categorize BUN (Blood Urea Nitrogen from LBDSBUSI)
def categorize_bun(val):
    if pd.isna(val):
        return np.nan
    elif val < 7:
        return "Low"
    elif val <= 20:
        return "Normal"
    else:
        return "High"

# Apply categorization
merged_df["BUN_Category"] = merged_df["LBDSBUSI"].apply(categorize_bun)
```

```python
# ---------------------------------------------------
# Step 1: Load RXQ_RX dataset (Prescription Medications)
# ---------------------------------------------------
rx_df = load_data("RXQ_RX", ["SEQN", "RXDDRUG"])

# Select only one of the SEQN columns if there are duplicates
if 'SEQN' in rx_df.columns and rx_df.columns.tolist().count('SEQN') > 1:
    rx_df = rx_df.loc[:, ~rx_df.columns.duplicated(keep='first')].copy()


# ---------------------------------------------------
# Step 2: Define keywords for CKD-related medications
# ---------------------------------------------------
ckd_drugs = [
    "lisinopril", "enalapril", "ramipril",           # ACE inhibitors
    "losartan", "valsartan", "irbesartan",           # ARBs
    "furosemide", "bumetanide", "torsemide",         # Loop diuretics
    "hydrochlorothiazide", "chlorthalidone",         # Thiazide diuretics
    "spironolactone", "eplerenone", "metolazone"     # Aldosterone antagonists
]

# ---------------------------------------------------
# Step 3: Flag prescriptions containing CKD drugs
# ---------------------------------------------------
# Decode byte strings and convert to lowercase for case-insensitive match
rx_df["RXDDRUG_lower"] = rx_df["RXDDRUG"].str.decode('utf-8').str.lower()


# Create boolean flag
rx_df["TakesCKDmed"] = rx_df["RXDDRUG_lower"].apply(
    lambda x: any(drug in str(x) for drug in ckd_drugs)
)

# ---------------------------------------------------
# Step 4: Collapse to 1 row per SEQN (any match = 1)
# ---------------------------------------------------
ckd_med_flags = rx_df.groupby("SEQN")["TakesCKDmed"].max().reset_index()

# ---------------------------------------------------
# Step 5: Merge with merged_df using a left merge
# ---------------------------------------------------
merged_df = merged_df.merge(ckd_med_flags, on="SEQN", how="left")

# Fill missing (i.e., no RX data for that participant) with 0
merged_df["TakesCKDmed"] = merged_df["TakesCKDmed"].fillna(0).astype(int)

# ---------------------------------------------------
# Step 6: Compute % taking CKD meds by NHANES cycle
# ---------------------------------------------------
ckd_med_percent = merged_df.groupby("YEAR")["TakesCKDmed"].mean() * 100

print("‚úÖ % Taking CKD-Related Medication by Cycle:")
print(ckd_med_percent.round(2))

# Line plot of % taking CKD-related medication by cycle
# ---------------------------------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
ckd_med_percent.sort_index().plot(kind="line", marker='o', linewidth=2, color="teal")
plt.title("% of Participants Taking CKD-Related Medications by NHANES Cycle")
plt.ylabel("Percentage (%)")
plt.xlabel("NHANES Cycle")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
