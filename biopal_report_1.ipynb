**uACR(urine albumin-creatinine ratio)**

Aids in early detection of CKD as it shows whether you have albumin in your urine. Albumin is an essential protein that's normally found in the blood; it helps build muscle, repair tissue, and fight infection. But it should be in your blood, not your urine. When you have albumin (protein) in your urine, it is called albuminuria or proteinuria. One of the main jobs of your kidneys is to filter your blood; they retain important biomolecules such as protein, whilst excreting waste products like creatinine. If kidneys are healthy, they should let only very little protein go into your urine, or even none. But if your kidneys are damaged, protein can "leak" out of the kidneys into your urine. High albumin in urine is an early indicator of CKD, which can ultimately progress to kidney failure if left untreated. (link)

**Uric Acid**

Another biomarker that indicates kidney dysfunction. Uric acid promotes inflammation, oxidative stress and endothelial dysfunction which can lead to damage of renal vasculature and glomeruli, impairing filtration of blood. (link) It can also precede a decline in eGFR levels making it possible for even earlier detection of CKD onset. (link) Uric acid is produced by the breakdown of purines in the liver and is supposed to be filtered out of blood by kidneys. If kidney function declines, uric acid is not filtered effectively leading to build up in the blood (hyperuricemia). Hyperuricemia may also promote CKD progression. (link)

**BUN (Blood Urea Nitrogen)**

Measures the amount of nitrogen in the blood from urea, a waste product formed from protein metabolism. Protein consumed is broken down into amino acids which are then broken down in the liver for energy or storage, and removing nitrogen from the amino acids. Nitrogen is converted into ammonia (toxic) that is detoxified in the liver by being converted to urea (water soluble & safe to circulate in the blood). Urea then enters the blood and BUN measures the nitrogen portion of the urea present. Kidneys are supposed to filter urea out of the blood into urine. When kidney filtration function decreases, BUN levels increase due to urea accumulating in blood. Normal BUN levels are 7–20 mg/dL but can differ slightly. Higher levels = kidney dysfunction, dehydration, high protein intake. Lower levels = liver disease, malnutrition, overhydration. (link)

- **Normality Test – Shapiro-Wilk**

To test if variables were normally distributed, the Shapiro-Wilk test was applied. However, this test requires that:

- Sample size must be between **3 and 5000**
- Variables must contain sufficient non-missing values

As some variables had more than 5000 valid samples, we randomly sampled **5000 rows** for each testable variable. Variables with zero valid samples were skipped and manually corrected.

```python
import pandas as pd

# Load all merged datasets
indep = pd.read_csv('/content/merged_independent_variables_2007_2018.csv')
urxuma = pd.read_csv('/content/merged_urxuma_2007_2018.csv')
urxucr = pd.read_csv('/content/merged_urxucr_2007_2018.csv')
urxuac = pd.read_csv('/content/merged_urxuac_2007_2018.csv')
bun = pd.read_csv('/content/merged_lbdsbusi_2007_2018.csv')

# Merge all datasets on SEQN
df = indep.merge(urxuma, on='SEQN', how='outer')
df = df.merge(urxucr, on='SEQN', how='outer')
df = df.merge(urxuac, on='SEQN', how='outer')
df = df.merge(bun, on='SEQN', how='outer')

# Filter to participants aged 20 and above
age_columns = [col for col in df.columns if "RIDAGEYR" in col]
df["RIDAGEYR"] = df[age_columns].bfill(axis=1).iloc[:, 0]
df = df[df["RIDAGEYR"] >= 20]

# Consolidate key variables using bfill to combine across years
df['URXUMA'] = df.filter(like='URXUMA').bfill(axis=1).iloc[:, 0]
df['URXUCR'] = df.filter(like='URXUCR').bfill(axis=1).iloc[:, 0]
df['URXUAC'] = df.filter(like='URXUAC').bfill(axis=1).iloc[:, 0]
df['LBDSBUSI'] = df.filter(like='LBDSBUSI').bfill(axis=1).iloc[:, 0]
df['BMXBMI'] = df.filter(like='BMXBMI').bfill(axis=1).iloc[:, 0]
df['INDFMIN2'] = df.filter(like='INDFMIN2').bfill(axis=1).iloc[:, 0]

# Save cleaned version
cleaned_path = "/content/cleaned_key_variables.csv"
df.to_csv(cleaned_path, index=False)

cleaned_path
```

```python
# 📦 Imports
import pandas as pd
from scipy.stats import shapiro

# 📁 Load cleaned dataset
df = pd.read_csv('/content/cleaned_key_variables.csv')

# 🎯 Variables to test (numerical)
numerical_vars = ['URXUMA', 'URXUCR', 'URXUAC', 'LBDSBUSI', 'RIDAGEYR', 'BMXBMI', 'INDFMIN2']

# 📊 Sample ≤ 5000 and drop NaNs for each variable
shapiro_results = []

for var in numerical_vars:
    data = df[var].dropna()
    if len(data) > 5000:
        data = data.sample(5000, random_state=42)
    if len(data) >= 3:  # Shapiro-Wilk requires at least 3 samples
        stat, p = shapiro(data)
        shapiro_results.append({'Variable': var, 'W-stat': stat, 'p-value': p})
    else:
        shapiro_results.append({'Variable': var, 'W-stat': None, 'p-value': None})

# 📄 Convert to DataFrame and display
shapiro_df = pd.DataFrame(shapiro_results)
print("📊 Shapiro-Wilk Test Results (≤5000 sample, no NaNs):")
print(shapiro_df)
```

**Interpretation:**  
All consolidated continuous variables showed p-values < 0.05 in the Shapiro-Wilk test, indicating significant deviation from normality, which is expected for biological data.
Thus, non-parametric statistical tests are more appropriate for subsequent analyses.

### 🔴 Percentage of Participants in High-Risk Biomarker Ranges

We define clinically abnormal values based on guidelines:

- **Albumin-to-Creatinine Ratio (uACR)** ≥ 30 mg/g indicates early-stage kidney damage.
- **Uric Acid (URXUAC)** ≥ 7.0 mg/dL is considered hyperuricemia.
- **Blood Urea Nitrogen (LBDSBUSI)** ≥ 20 mg/dL may indicate renal impairment or dehydration.

The following plot shows the percentage of individuals in each NHANES cycle who fall into these risk groups.


```python
# Define cycle order
cycle_order = ['2007_2008', '2009_2010', '2013_2014', '2015_2016', '2017_2018']

# Reassign Cycle column
def detect_cycle(row):
    for cycle in cycle_order:
        if any(pd.notna(row[col]) for col in df.columns if f"_{cycle}" in col):
            return cycle
    return pd.NA

df['Cycle'] = df.apply(detect_cycle, axis=1)

# Compute uACR in mg/g (URXUMA in mg/L, URXUCR in mg/dL)
df['uACR'] = (df['URXUMA'] / df['URXUCR']) * 100

# ⬆️ Define clinical high-risk thresholds
df['High_uACR'] = df['uACR'] >= 30      # uACR ≥ 30 mg/g (CKD threshold)
df['High_URXUAC'] = df['URXUAC'] >= 7.0    # arbitrary example for hyperuricemia
df['High_BUN'] = df['LBDSBUSI'] > 20    # BUN > 20 mg/dL may indicate poor kidney function



# 🧮 Compute percentage of high-risk cases per cycle
high_risk_trends = (
    df.dropna(subset=['Cycle'])  # exclude missing cycles
    .groupby('Cycle')[['High_uACR', 'High_URXUAC', 'High_BUN']]
    .mean() * 100  # convert to percentages
).round(2)

display(high_risk_trends)
import matplotlib.pyplot as plt

# Create individual plots
fig, axs = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

# uACR
axs[0].plot(high_risk_trends.index, high_risk_trends['High_uACR'], marker='o')
axs[0].set_title('High-Risk uACR (≥30 mg/g) by Cycle')
axs[0].set_ylabel('Percentage (%)')
axs[0].grid(True)

# URXUAC
axs[1].plot(high_risk_trends.index, high_risk_trends['High_URXUAC'], marker='o', color='green')
axs[1].set_title('High-Risk Uric Acid (>7 mg/dL) by Cycle')
axs[1].set_ylabel('Percentage (%)')
axs[1].grid(True)

# BUN
axs[2].plot(high_risk_trends.index, high_risk_trends['High_BUN'], marker='o', color='red')
axs[2].set_title('High-Risk BUN (>20 mg/dL) by Cycle')
axs[2].set_ylabel('Percentage (%)')
axs[2].set_xlabel('NHANES Cycle')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

**Intepretation:**
- Percentage of individuals with High-Risk uACR shows a notable drop from about 13.2% to 10.3% after 2008 and gradually rises till 12.9% from 2010 to 2018
- The trendline is consistent to the percentage of individuals diagnosed with CKD as mentioned before.
- Percentage of individuals with High-Risk uric acid increased significantly from 0.020% to 0.060% after 2008 and gradually dropped till 0.020% from 2010 to 2018
- Percentage of individuals with High-Risk BUN did not change from 2007-2010 and stayed constant at 0.10%. A significant increase from 0.10% to 0.19% is observed from 2010-2014. A slight drop to 0.16% is then observed from 2014-2018

These temporal patterns highlight how underlying factors can influence biomarkers overtime
